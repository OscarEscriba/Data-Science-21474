```python
import os
import requests

# Lista de URLs proporcionadas
zip_urls = [
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/87242/importacion_de_vehiculos_2025_mayo.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/86197/importacion_de_vehiculos_2025_abril.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/85782/importacion_de_vehiculos_2025_marzo.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/85486/importacion_de_vehiculos_2025_febrero.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/84293/importacion_de_vehiculos_2025_enero.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/83292/importacion_de_vehiculos_2024_diciembre.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/82785/importacion_de_vehiculos_2024_noviembre.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/82449/importacion_de_vehiculos_2024_octubre.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/81634/importacion_de_vehiculos_2024_septiembre.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/81358/importacion_de_vehiculos_2024_agosto.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/80559/importacion_de_vehiculos_2024_julio.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/79819/importacion_de_vehiculos_2024_junio.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/79093/importacion_de_vehiculos_2024_mayo.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/78234/importacion_de_vehiculos_2024_abril.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/77313/importacion_de_vehiculos_2024_marzo.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/76897/importacion_de_vehiculos_2024_febrero.zip",
    "https://portal.sat.gob.gt/portal/descarga/5030/importacion-de-vehiculos/75808/importacion_de_vehiculos_2024_enero.zip"
]

# Crear carpeta si no existe
os.makedirs("Zips", exist_ok=True)

# Descargar los archivos
for url in zip_urls:
    filename = url.split("/")[-1]
    filepath = os.path.join("Zips", filename)
    if not os.path.exists(filepath):
        print(f"Descargando {filename}...")
        r = requests.get(url)
        with open(filepath, "wb") as f:
            f.write(r.content)
    else:
        print(f"{filename} ya existe, omitiendo.")

print("‚úÖ Todos los archivos han sido descargados.")
```

```python
# importar los archivos ZIP y guardarlos en una carpeta llamada "datos"
import zipfile

# Crear carpeta de destino
os.makedirs("Datos", exist_ok=True)

# Listar archivos .zip descargados
zip_folder = "Zips"
zip_files = [f for f in os.listdir(zip_folder) if f.endswith(".zip")]

# Descomprimir cada archivo
for zip_name in zip_files:
    zip_path = os.path.join(zip_folder, zip_name)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        print(f"Descomprimiendo: {zip_name}")
        zip_ref.extractall("Datos")

print("‚úÖ Todos los archivos han sido descomprimidos en la carpeta 'Datos'.")
```

```python
# codigo para leet todos los txt y unirlos como un solo dataframe
import pandas as pd

# Definir columnas esperadas (seg√∫n lo que mencionaste)
columnas = [
    "Pais de Proveniencia", "Aduana de Ingreso", "Fecha de la Poliza", "Partida Arancelaria",
    "Modelo del Vehiculo", "Marca", "Linea", "Centimetros Cubicos", "Distintivo",
    "Tipo de Vehiculo", "Tipo de Importador", "Tipo Combustible", "Asientos",
    "Puertas", "Tonelaje", "Valor CIF", "Impuesto"
]

# Leer todos los .txt en la carpeta "Datos"
txt_folder = "Datos"
archivos_txt = [f for f in os.listdir(txt_folder) if f.endswith(".txt")]

# Lista para almacenar los DataFrames
df_list = []

# Leer uno por uno (versi√≥n mejorada que ignora columna vac√≠a extra)
for archivo in archivos_txt:
    path = os.path.join(txt_folder, archivo)
    print(f"Leyendo: {archivo}")
    try:
        df = pd.read_csv(
            path,
            sep="|",
            encoding="latin-1",
            names=columnas,
            header=None,
            engine="python",
            usecols=range(17)  # leer solo las primeras 17 columnas
        )
        df_list.append(df)
    except Exception as e:
        print(f"‚ùå Error en archivo {archivo}: {e}")

# Unir todos los DataFrames en uno solo
df_final = pd.concat(df_list, ignore_index=True)

print(f"\n‚úÖ Archivos le√≠dos: {len(df_list)}")
print(f"üî¢ Total de filas combinadas: {len(df_final)}")
```

```python
# Ver cu√°ntos duplicados hay
duplicados = df_final.duplicated().sum()
print(f"üîÅ Registros duplicados encontrados: {duplicados}")

# Eliminar duplicados
df_final = df_final.drop_duplicates()

print(f"‚úÖ Total de registros √∫nicos despu√©s de limpieza: {len(df_final)}")
```

```python
# Guardar el DataFrame limpio
df_final.to_csv("importaciones_vehiculos_2024_2025.csv", index=False, encoding="utf-8")
print("‚úÖ Archivo limpio guardado como 'importaciones_vehiculos_2024_2025.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Asegurar que las fechas est√©n en formato datetime
df_final["Fecha de la Poliza"] = pd.to_datetime(df_final["Fecha de la Poliza"], dayfirst=True, errors="coerce")

# Filtrar registros v√°lidos
df_final = df_final.dropna(subset=["Fecha de la Poliza"])

# Agregar columnas auxiliares
df_final["A√±o"] = df_final["Fecha de la Poliza"].dt.year
df_final["Mes"] = df_final["Fecha de la Poliza"].dt.month

# Filtrar por veh√≠culos livianos
livianos = df_final[df_final["Distintivo"].str.upper().str.contains("LIVIANO", na=False)]

# 1. ¬øCu√°ntos veh√≠culos livianos de cada tipo se importaron en 2024?
tipos_livianos_2024 = livianos[livianos["A√±o"] == 2024]["Tipo de Vehiculo"].value_counts()

# 2. ¬øDistribuci√≥n de modelos de carros, pickups y SUV que se importaron el a√±o pasado?
filtro_vehiculos = df_final[
    (df_final["A√±o"] == 2024) & 
    (df_final["Tipo de Vehiculo"].str.upper().isin(["CARRO", "PICK UP", "PICKUP", "SUV"]))
]
modelos_distribucion = filtro_vehiculos.groupby("Tipo de Vehiculo")["Modelo del Vehiculo"].value_counts()

# 3. ¬øCu√°l es el tipo de veh√≠culo que m√°s se import√≥ en 2024?
tipo_mas_importado = df_final[df_final["A√±o"] == 2024]["Tipo de Vehiculo"].value_counts().idxmax()

# 4. ¬øMeses en los que m√°s se importan veh√≠culos livianos?
meses_livianos_2024 = livianos[livianos["A√±o"] == 2024]["Mes"].value_counts().sort_index()

# 5. Comparaci√≥n a√±o a a√±o por tipo de veh√≠culo (mes a mes)
df_comparativo = df_final[df_final["Tipo de Vehiculo"].notna()]
importaciones_comparativas = df_comparativo.groupby(["A√±o", "Mes", "Tipo de Vehiculo"]).size().reset_index(name="Cantidad")

# Mostrar resultados
print("1Ô∏è‚É£ Veh√≠culos livianos de cada tipo en 2024:\n", tipos_livianos_2024, "\n")
print("2Ô∏è‚É£ Distribuci√≥n de modelos de Carros, Pickups y SUV en 2024:\n", modelos_distribucion, "\n")
print(f"3Ô∏è‚É£ Tipo de veh√≠culo m√°s importado en 2024: {tipo_mas_importado}\n")
print("4Ô∏è‚É£ Meses con m√°s importaci√≥n de veh√≠culos livianos en 2024:\n", meses_livianos_2024, "\n")

# 5Ô∏è‚É£ Gr√°fica de comparaci√≥n mensual por tipo de veh√≠culo 2024 vs 2025
pivot_comp = importaciones_comparativas.pivot_table(
    index=["Mes", "Tipo de Vehiculo"],
    columns="A√±o",
    values="Cantidad",
    fill_value=0
).reset_index()

# Mostrar comparaci√≥n para veh√≠culos livianos
pivot_livianos = pivot_comp[pivot_comp["Tipo de Vehiculo"].isin(livianos["Tipo de Vehiculo"].unique())]

print("5Ô∏è‚É£ Comparaci√≥n mensual por tipo de veh√≠culo (2024 vs 2025):")
display(pivot_livianos)

# Opcional: mostrar gr√°ficas
plt.figure(figsize=(10,5))
meses_livianos_2024.plot(kind="bar", color="skyblue")
plt.title("Importaci√≥n mensual de veh√≠culos livianos en 2024")
plt.xlabel("Mes")
plt.ylabel("Cantidad")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
